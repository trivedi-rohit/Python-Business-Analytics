```python
marks_sum = 0
for i in marks:
  marks_sum = i + marks_sum

marks_mean = marks_sum/len(marks)
print(marks_mean)
```

```python
def my_mean(marks):
  marks_sum = 0
  for i in marks:
    marks_sum = marks_sum + i

  marks_mean = marks_sum/len(marks)
  return marks_mean
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x,y)
plt.show()
```

```python
m, c, r, p, std_err = stats.linregress(x,y)  #from stats, we are using linear regression

def myfun(x):
  return m*x + c

mymodel = list(map(myfun,x))  #runiing each value of x array through myfun. This is generating new list with new values for y-axis

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show
```

```python
predict_y = myfun(14)
print(predict_y)
```
